```python
import simpy
import random 
import numpy as np
```

```python
class JusticeSystem():
    def __init__(self, env) -> None:
        self.active_cases = 0
        self.env = env

        # Some random, hard-coded constants
        self.summoned_solved_prob = 0.3


    # The summoned process
    def summoned(self, env):
        while True:
            # Influx of cases
            self.active_cases += 1

            # Cases solved
            if np.random.uniform() < self.summoned_solved_prob:
                self.active_cases -= 1
                yield self.env.timeout(1)
            
            else: yield self.env.timeout(1)
            
            # State at the end of the day
            print(f'At the end of day {env.now}, the number of active cases is {self.active_cases},000 cases')
```

```python
def simulate(env):
    justice_system = JusticeSystem(env)
    env.process(justice_system.summoned(env)) #Add process `summoned` to the simpy env
    yield env.timeout(0)
```

```python
env = simpy.Environment()

env.process(simulate(env)) # Add process `simulate` to the simpy env
env.run(until=100)
```
