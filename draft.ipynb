```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 random numbers from a uniform distribution between 4 and 5
data = np.random.normal(5, 0.4, 1000)

# Plot the histogram
plt.hist(data, bins=50, density=True, alpha=0.7, color='g')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.title('Uniform Distribution between 4 and 5')
plt.show()
```
